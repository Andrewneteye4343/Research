```python
#####################################################################################################################
# Code for AutoDock Vina
# Import modules
import subprocess
import os

# Input path to the parent directory as "path" variable.
# You can use the command "pwd" in linux to access the current path.
path = "/path/to/Docking/"
# Define a list, named "drugs", comprised with yours sub-directory names (often be the compound/drug names).
# These elements in "drugs" list are the directroy names that you created with mkdir.py before.
drugs = ["Compound1_directory","Compound2_directory", "Compound3_directory", ...] 

# Define how many times that the molecular docking will be calculated.
time_docking = 1

# Define the "docking" function
# By using "subprocess", we can write linux commands with python codes.
# Here, AutoDock Vina will be operated with only one input conf.txt, which is written with AutoDock Vina parameters (e.g. receptor, ligand, center_x, size_x...)
# This "docking" function will be operated in 1 sub-directory once per time. After that, it automatically turns to another sub-directory, and operate.

def docking():
    a = subprocess.run("vina --config conf.txt", shell=True)
    if a.returncode != 0:
        print("Returncode is", a.returncode)
    else:
        print("Docking successfully!")

for drug in drugs:
    workpath = path + str(drug) + "/"
    os.chdir(workpath)
    print("This is in ", drug)
    for i in range(1, time_docking+1):
        docking()
        number = "0"+str(i)
        b = subprocess.run(["mv", "log.txt", "log"+number+".txt"])
        c = subprocess.run(
            ["mv", "ligand_out.pdbqt", "ligand_out"+number+".pdbqt"])
        if b.returncode != 0:
            print("Returncode is", b.returncode)
        else:
            print("Rename log file successfully!")
        if c.returncode != 0:
            print("Returncode is", c.returncode)
        else:
            print("Rename output file successfully!")

#####################################################################################################################
# Content of conf.txt
receptor = /path/to/Docking/receptor.pdbqt
ligand = ligand.pdbqt
size_x = 84
size_y = 60
size_z = 62
center_x = 64.124
center_y = 104.772
center_z = 89.486
exhaustiveness = 48
log = log.txt
#####################################################################################################################
```
